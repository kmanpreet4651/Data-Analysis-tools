```python
import pandas as pd
import matplotlib.pyplot as plt

# List of categorical columns to visualize
categorical_columns = ['Gender', 'Item Purchased', 'Category', 'Payment Method', 'Shipping Type']

# Create bar plots and pie charts
for column in categorical_columns:
    # Bar plot
    plt.figure(figsize=(8, 4))
    df[column].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Bar Plot of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()

    # Pie chart
    plt.figure(figsize=(6, 6))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
    plt.title(f'Pie Chart of {column}', fontsize=14)
    plt.ylabel('')  # Hide the y-axis label for a cleaner pie chart
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram: Distribution of Numerical Data
plt.figure(figsize=(8, 4))
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df['Purchase Amount (USD)'], bins=5, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Distribution of Purchase Amount', fontsize=14)
plt.xlabel('Purchase Amount (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 2. Box Plot: Highlight Outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x='Category', y='Review Rating', data=df, palette='pastel')
plt.title('Review Ratings by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Review Rating', fontsize=12)
plt.tight_layout()
plt.show()

# 3. Bar Graph: Frequencies (Items Purchased by Gender)
plt.figure(figsize=(8, 4))
item_gender_counts = df.groupby(['Gender', 'Item Purchased']).size().unstack()
item_gender_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(8, 4))
plt.title('Items Purchased by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count of Items Purchased', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Items Purchased', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Detect Seasonal Trends
plt.figure(figsize=(8, 4))
season_item_trends = df.groupby(['Season', 'Category']).size().unstack()
season_item_trends.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.title('Seasonal Trends of Category', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Count of Category', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
season_color_trends = df.groupby(['Season', 'Color']).size().unstack()
season_color_trends.plot(kind='bar', stacked=True, colormap='plasma', figsize=(10, 6))
plt.title('Seasonal Trends of Colors', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Count of Colors', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Colors', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 2. Spot Anomalies in Review Ratings and Purchase Amount
# Box plot for anomalies in Review Ratings
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Review Rating'], color='lightblue')
plt.title('Anomalies in Review Ratings', fontsize=14)
plt.xlabel('Review Rating', fontsize=12)
plt.tight_layout()
plt.show()

# Box plot for anomalies in Purchase Amount
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Purchase Amount (USD)'], color='salmon')
plt.title('Anomalies in Purchase Amount', fontsize=14)
plt.xlabel('Purchase Amount (USD)', fontsize=12)
plt.tight_layout()
plt.show()

# Highlight Anomalies (Values outside the IQR range)
def detect_anomalies(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    anomalies = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return anomalies

review_rating_anomalies = detect_anomalies('Review Rating')
purchase_amount_anomalies = detect_anomalies('Purchase Amount')

print("Anomalies in Review Ratings:")
print(review_rating_anomalies)

print("\nAnomalies in Purchase Amount:")
print(purchase_amount_anomalies)
```
